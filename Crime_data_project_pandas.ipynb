```python
#Removing duplicates
df = df.drop_duplicates()
```

```python
#Dropping unwanted coloumns
df.drop(columns=['Crime code 3'],inplace=True)
df.drop(columns=['Crime code 4'],inplace=True)
```

```python
#ENSURE THE COLOUMN AREA ONLY CONTAINS INTEGERS
df['Area']=df['Area'].astype('int64')
df['Premise code']=df['Premise code'].astype('int64')
df['Weapon code']=df['Weapon code'].astype('int64')
```

```python
#Standardizing Categorical Data
df['Crime description'] = df['Crime description'].str.lower().str.strip()
```

```python
#Ensure correct data types for each column
df['Latitude'] = df['Latitude'].astype(float)
df['Longitude'] = df['Longitude'].astype(float)
```

```python
#Ensure latitude and longitude are within valid ranges
df = df[(df['Latitude'].between(-90, 90)) & (df['Longitude'].between(-180, 180))]
```

```python
#Changing format of Date reported
df['Date reported']=pd.to_datetime(df['Date reported'],format='%d-%m-%Y')
```

ADVANCED CRIME DATA ANALYSIS

#### 1_RELATIONSHIP BETWEEN THE AGE OF VICTIM AND THE TYPE OF CRIME THEY COMMITED

```python

import pandas as pd
import matplotlib.pyplot as plt
age_categories = pd.cut(df['Victim age'], bins=[0, 18, 25, 35, 65, 120])

# Group data by age category and crime code
crime_counts_by_age = df.groupby([age_categories, 'Crime description'])['Crime description'].count().unstack()
crime_counts_by_age.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Age Category')
plt.ylabel('Crime Count')
plt.title('Crime Type Distribution by Age Category')
plt.xticks(rotation=45)
plt.show()

import seaborn as sns

sns.heatmap(crime_counts_by_age, annot=True, fmt='d')  
plt.xlabel('Crime description')
plt.ylabel('Age category')
plt.title('Crime Type Distribution by Age Category (Heatmap)')
plt.show()
```

The age group between 35 and 65 years is associated with the highest number of crimes, including offenses like brandishing weapons, rape, and robbery.

#### 2_GEOGRAPHICAL DISTRIBUTION OF CRIMES

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Longitude', y='Latitude', data=df, hue='Area name', legend=False)
plt.title('Geographical Distribution of Crimes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

 Crimes are distributed randomly all over without any any specific trend.

#### 3_Identifying total Crime description,Weapon description in different Areas

```python
crime_counts = df.groupby('Area')['Crime description'].count().reset_index(name='Total Crimes')
weapon_counts = df.groupby('Area')['Weapon description'].count().reset_index(name='Total Weapons')
area_counts = pd.merge(crime_counts, weapon_counts, on='Area')
print(area_counts)


plt.figure(figsize=(14, 7))
sns.barplot(x='Area', y='Total Crimes', data=area_counts, palette='viridis')
plt.title('Total Crimes by Area')
plt.xlabel('Area')
plt.ylabel('Total Crimes')
plt.xticks(rotation=90)
plt.show()

plt.figure(figsize=(14, 7))
sns.barplot(x='Area', y='Total Weapons', data=area_counts, palette='magma')
plt.title('Total Weapon Descriptions by Area')
plt.xlabel('Area')
plt.ylabel('Total Weapons')
plt.xticks(rotation=90)
plt.show()
```

Area '6' has the highest number of crimes and weapon usage.

#### 4_The most common crimes

```python

common_crimes = df['Crime description'].value_counts()
print(common_crimes)
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.barplot(x=common_crimes.values, y=common_crimes.index, palette='viridis')
plt.title('Top 10 Most Common Crimes')
plt.xlabel('Number of occurence')
plt.ylabel('Crime description')
plt.show()
```

Robbery is the most common type of crime.
The age group  between 25 to 50 has done the most crimes with strong arms,in that robbery is the common type of crime.

#### 5_Areas with the Highest Crime Rates

```python

high_crime_areas = df['Area name'].value_counts().head(10)
print(high_crime_areas)
import matplotlib.pyplot as plt
import seaborn as sns

area_counts = df.groupby('Area name').size()
plt.figure(figsize=(10, 8))
area_counts.plot.pie(autopct='%1.1f%%', startangle=140)
plt.title('Crime Reports Distribution by Area Name')
plt.ylabel('') 
plt.show()
```

Hollywood has the highest crime rate distribution compared to other areas.

## Crime Data Analysis Project Summary

In this project, we conducted an extensive analysis of crime data to uncover patterns and insights related to victim age, types of crimes committed, weapons used, and geographical distribution. The key findings from this analysis are summarized below

### Age Group Analysis:

The age group between 35 and 65 years is associated with the highest number of crimes. These crimes include serious offenses such as brandishing weapons, rape, and robbery.
Within the age group of 25 to 50 years, there is a significant occurrence of crimes committed with strong arms, with robbery being the most common type of crime in this category.

### Crime Distribution:

Crimes are distributed randomly across different areas, without any specific trend that can be discerned from the data. This suggests that crimes are not concentrated in particular locations based on any clear pattern.
Area Analysis:

Area '6' has the highest number of reported crimes as well as the highest usage of weapons. This indicates a hotspot for criminal activity within the dataset.
Hollywood stands out as the area with the highest crime rate distribution compared to other regions analyzed.

### Crime Status:

When examining the status description of crimes, it was found that the status 'Invest' has the highest count. This indicates a significant number of ongoing investigations related to the reported crimes.

The findings from this analysis provide valuable insights into the distribution and characteristics of crimes within the dataset. These insights can be useful for law enforcement and policymakers to better understand crime dynamics and allocate resources effectively.
